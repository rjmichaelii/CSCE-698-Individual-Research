```python
import os

import numpy as np
import pandas as pd
```

# A-novel-statistical-physical_Usowicz and Lipiec, 2026

```python
df_1 = pd.read_csv(f'{raw_path}/F3a_Heat_Capacity_vs_Water_Content.csv')
```

```python
df_1.head()
```

```python
df_1.columns = df_1.columns.str.strip()
```

Convert volumetric heat capacity (MJ/m^3/K) to specific heat (J/kg/K). With density in Mg/m^3, the unit conversion is 1000 * Cv / rho.

```python
df_1['Specific_Heat_(Jkg^-1K^-1)'] = 1000 * (df_1['Heat_Capacity_(MJm^-3K^-1)'] / df_1['Bulk_Density_(Mgm^-3)'])
```

```python
df_1.head()
```

```python
new_rows = df_1[['Water_Content_(m^3m^-3)', 'Specific_Heat_(Jkg^-1K^-1)', 'Bulk_Density_(Mgm^-3)']].rename(columns={
    'Water_Content_(m^3m^-3)': 'Water_Content',
    'Specific_Heat_(Jkg^-1K^-1)': 'Specific_Heat',
    'Bulk_Density_(Mgm^-3)': 'Bulk_Density'
})
new_rows['Data_Type'] = 'Observed'
new_rows['Soil'] = np.nan
new_rows['Source'] = 'Usowicz and Lipiec 2026'
new_rows['Water_Content_Basis'] = 'volumetric'
new_rows = new_rows[['Water_Content', 'Specific_Heat', 'Bulk_Density', 'Data_Type', 'Soil', 'Source', 'Water_Content_Basis']]

specific_heat_df = pd.concat([specific_heat_df, new_rows], ignore_index=True)
```

```python
specific_heat_df.shape
```

```python
specific_heat_df.tail()
```

# TheramlProperties_Nidal H. Abu-Hamdeh_2003

```python
df_6 = pd.read_csv(f'{raw_path}/F1_Clay_Soil_Specific_Heat.csv')
df_6.columns = df_6.columns.str.strip()
df_6['Moisture_Content_w_(kg/kg)'] = pd.to_numeric(df_6['Moisture_Content_w_(kg/kg)'], errors='coerce')
df_6['Specific_Heat_c_(kJ/kg)'] = pd.to_numeric(df_6['Specific_Heat_c_(kJ/kg)'], errors='coerce')
df_6['Data_Type'] = df_6['Data_Type'].str.strip()
df_6['Specific_Heat_(Jkg^-1K^-1)'] = df_6['Specific_Heat_c_(kJ/kg)'] * 1000

new_rows = df_6.rename(columns={
    'Moisture_Content_w_(kg/kg)': 'Water_Content',
    'Specific_Heat_(Jkg^-1K^-1)': 'Specific_Heat'
})
new_rows['Soil'] = 'Clay'
new_rows['Source'] = 'Abu-Hamdeh 2003'
new_rows['Water_Content_Basis'] = 'gravimetric'
new_rows = new_rows[['Water_Content', 'Specific_Heat', 'Data_Type', 'Soil', 'Source', 'Water_Content_Basis']]

specific_heat_df = pd.concat([specific_heat_df, new_rows], ignore_index=True)
specific_heat_df.tail()
```

```python
df_7 = pd.read_csv(f'{raw_path}/F2_Sandy_Soil_Specific_Heat.csv')
df_7.columns = df_7.columns.str.strip()
df_7['Moisture_Content_w_(kg/kg)'] = pd.to_numeric(df_7['Moisture_Content_w_(kg/kg)'], errors='coerce')
df_7['Specific_Heat_(kJ/kg)'] = pd.to_numeric(df_7['Specific_Heat_(kJ/kg)'], errors='coerce')
df_7['Data_Type'] = df_7['Data_Type'].str.strip()
df_7['Specific_Heat_(Jkg^-1K^-1)'] = df_7['Specific_Heat_(kJ/kg)'] * 1000

new_rows = df_7.rename(columns={
    'Moisture_Content_w_(kg/kg)': 'Water_Content',
    'Specific_Heat_(Jkg^-1K^-1)': 'Specific_Heat'
})
new_rows['Soil'] = 'Sand'
new_rows['Source'] = 'Abu-Hamdeh 2003'
new_rows['Water_Content_Basis'] = 'gravimetric'
new_rows = new_rows[['Water_Content', 'Specific_Heat', 'Data_Type', 'Soil', 'Source', 'Water_Content_Basis']]

specific_heat_df = pd.concat([specific_heat_df, new_rows], ignore_index=True)
specific_heat_df.tail()
```

## Standardize For Modeling

```python
specific_heat_df = specific_heat_df.copy()
for column in ['Water_Content', 'Specific_Heat', 'Bulk_Density', 'Data_Type', 'Soil', 'Source', 'Water_Content_Basis']:
    if column not in specific_heat_df.columns:
        specific_heat_df[column] = np.nan

specific_heat_df['Data_Type'] = specific_heat_df['Data_Type'].fillna('Observed')
specific_heat_df['Soil'] = specific_heat_df['Soil'].fillna('Unknown')

cp_moisture_model_df = specific_heat_df.rename(columns={
    'Specific_Heat': 'target_value',
    'Water_Content': 'water_content',
    'Bulk_Density': 'bulk_density',
    'Water_Content_Basis': 'water_content_basis',
    'Data_Type': 'data_type',
    'Soil': 'soil_class',
    'Source': 'source'
}).copy()
cp_moisture_model_df['temperature_c'] = np.nan
cp_moisture_model_df['dataset_family'] = 'moisture_based'
cp_moisture_model_df['target_type'] = 'specific_heat'
cp_moisture_model_df = cp_moisture_model_df[[
    'target_value', 'target_type', 'dataset_family',
    'water_content', 'water_content_basis', 'temperature_c', 'bulk_density',
    'soil_class', 'data_type', 'source'
]]

cp_moisture_model_df.head()
```

```python
cp_moisture_volumetric_df = cp_moisture_model_df[
    cp_moisture_model_df['water_content_basis'] == 'volumetric'
].reset_index(drop=True)
cp_moisture_gravimetric_df = cp_moisture_model_df[
    cp_moisture_model_df['water_content_basis'] == 'gravimetric'
].reset_index(drop=True)
cp_moisture_observed_df = cp_moisture_model_df[
    cp_moisture_model_df['data_type'] == 'Observed'
].reset_index(drop=True)
cp_moisture_volumetric_observed_df = cp_moisture_model_df[
    (cp_moisture_model_df['water_content_basis'] == 'volumetric')
    & (cp_moisture_model_df['data_type'] == 'Observed')
].reset_index(drop=True)

print('all moisture-based cp rows:', cp_moisture_model_df.shape)
print('volumetric-water rows:', cp_moisture_volumetric_df.shape)
print('gravimetric-water rows:', cp_moisture_gravimetric_df.shape)
print('observed rows:', cp_moisture_observed_df.shape)
print('volumetric observed rows:', cp_moisture_volumetric_observed_df.shape)
```

```python
os.makedirs(processed_path, exist_ok=True)
specific_heat_df.to_csv(f'{processed_path}/specific_heat_cleaned.csv', index=False)
cp_moisture_model_df.to_csv(f'{processed_path}/cp_moisture_model_df.csv', index=False)
cp_moisture_volumetric_df.to_csv(f'{processed_path}/cp_moisture_volumetric_df.csv', index=False)
cp_moisture_gravimetric_df.to_csv(f'{processed_path}/cp_moisture_gravimetric_df.csv', index=False)
cp_moisture_observed_df.to_csv(f'{processed_path}/cp_moisture_observed_df.csv', index=False)
cp_moisture_volumetric_observed_df.to_csv(
    f'{processed_path}/cp_moisture_volumetric_observed_df.csv',
    index=False,
)
```
